```python
def illustrate(m):
    import networkx as nx
    g = nx.DiGraph()
    nodes = m
    edges = [(k,v) for k,v in E.gen_rotor(E.chars).items()]
    g.add_nodes_from(nodes)
    g.add_edges_from(edges)
    nx.draw(g)
```
